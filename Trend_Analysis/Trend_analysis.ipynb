```python
df5 = {}
for i in range(len(df_trnd['Name'])):
    if 'Major League Soccer (MLS)' in df_trnd['Name'][i]:
        df5[i] =df_trnd.iloc[i,:]
df5
```

```python
x = []
for key,items in df5.items():
    x.append(items)
```

```python
from pylab import *
import numpy as np
import matplotlib.pyplot as plt
#import matplotlib.pyplot as plt; plt.rcdefaults()

for i in range(0,6):
    t = ['Jan','Feb','Mar','Apr','May','Jun','Jul','Aug','Sep','Oct','Nov','Dec']
    s = list(x[i][2:14])
    plt.bar(t,s, align='center', alpha=0.5)
    plt.plot(t,s)
    plt.ylabel(f'{x[i][0]}')
    plt.xlabel('Month')
    plt.show()
```

<b>Trend in number of Posts per month for MLS:</b>
1.	Here according to the graph, we can surely see that there is overall variation in number of posts over the year, but number of posts increased in the period from July to October. Maximum number of posts were for the period of August. 

<b>Trend in % Shares Fb per month for MLS:</b>

1.	According to the graph we can surely conclude that over all percentage of shares was lower in the span of year. However, there was sharp increase in percentage of shares per fb posts for the month of August and September followed by the sharp decline.

<b>Trend in Actions per post FB:</b>
1.	As it can be seen in the graph that after a small fall from January to February, there was an increase in number in March. After March the numbers are constant and again there is sharp increase for the period of September. After September again a decline in graph line can be seen.

<b>Trend in % Returning Engagers FB:</b>

1.	Overall the percentage or returning engagers of FB be constant throughout the year except the starting of the month.

<b>Trend in Unique Audience FB:</b>

1.	We can say that for most of the time during year, number of unique audiences on FB was very low. For the month of march and august, slight increase can be observed. For the month of September number of unique audiences increases sharply and after which number dropped drastically and kept decreasing.
