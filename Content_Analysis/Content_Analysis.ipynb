<b>Understanding Data columns</b>

1.	Page: Names of different Facebook pages, for which data is collected.
2.	Fans: Total number of fans for each page.
3.	Category: Category of what name of the page signifies. For example, MLS is of the category sports league
4.	Total Actions:  Summation of Reaction count, comments count and share count.
5.	Total Actions % Fans:  Percentage of total actions per total number of fans.
6.	Word count: Word count for the message:
7.	Post ID:  Unique post id for the post.
8.	Message: Heading for the post.
9.	Link:  Actual link to the post.
10.	Link to likes: Link of the Facebook posts that may redirect to the actual posts.
11.	Post Type: Type of the fb post, whether it is video or link or Photo.
12.	Promoted: True of False, whether post was promoted or not.
13.	Likes count: Total number of likes on the Facebook post (it does not include other reactions).
14.	Reaction count: Total number of reactions on the post (including likes, comments, and other reactions on the post and replies on post)
15.	Comments count:  Total comment threads on the post.
16.	Share count:  Total number of shares of the post.
17.	Lifetime video Views: Total views on the videos.
18.	Created time: Includes created time of the post.
19.	Promoted: True or false value. If post was promoted or not
20.	Call to like:  True or false value
21.	Call to share: True or false value
22.	Call to Website: True or false value
23.	Includes Question: True or false value
24.	Public Clicks: Number of public clicks who are from other sources not facebook users.
25. Destination Urls:  Actual page link to the post

<B>PLOTTING GRAPH FOR TOTAL ACTIONS FOR ALL 5 POST TYPES FOR EVERY HOUR</b>

<b>Analysis</b> 
Not much relation between creation time of post and post .This analysis was done to find whether posting a post during specific hour of the day has effect on the popularity of the post. And we found out the post.

```python
# PLOTTING GRAPH FOR TOTAL ACTIONS FOR ALL 5 POST TYPES FOR EVERY HOUR
cond1 =(df_mls.post_type == 'video')
cond2 =(df_mls.post_type == 'photo')
cond3 =(df_mls.post_type == 'link')
cond4 =(df_mls.post_type == 'status')
cond5 =(df_mls.post_type == 'event')

subset_a = df_mls[cond1]
subset_b = df_mls[cond2]
subset_c = df_mls[cond3]
subset_d = df_mls[cond4]
subset_e = df_mls[cond5]

plt.scatter(subset_a.hour, subset_a.total_actions, s=120, c='b', label='video')
plt.scatter(subset_b.hour, subset_b.total_actions, s=120, c='r', label='photo') 
plt.scatter(subset_c.hour, subset_c.total_actions, s=120, c='g', label='links')
plt.scatter(subset_d.hour, subset_d.total_actions, s=120, c='y', label='status')
plt.scatter(subset_e.hour, subset_e.total_actions, s=120, c='purple', label='event') 
plt.xlabel('No of hours')
plt.ylabel('Total actions')
plt.legend()


```

```python
# PLOTTING GRAPH FOR TOTAL ACTIONS FOR ALL 5 POST TYPES FOR EVERY HOUR
cond1 =(df_mls.post_type == 'video')
cond2 =(df_mls.post_type == 'photo')
cond3 =(df_mls.post_type == 'link')
cond4 =(df_mls.post_type == 'status')
cond5 =(df_mls.post_type == 'event')

summation =[]
x= []

subset_a = df_mls[cond1]
a=subset_a['total_actions']
temp=a.sum()
summation.append(temp)
temp2= subset_a['post_type'].unique()
x.append(temp2)

subset_b = df_mls[cond2]
b=subset_b['total_actions']
temp=b.sum()
summation.append(temp)
temp2= subset_b['post_type'].unique()
x.append(temp2)

subset_c = df_mls[cond3]
c=subset_c['total_actions']
temp=c.sum()
summation.append(temp)
temp2= subset_c['post_type'].unique()
x.append(temp2)

subset_d = df_mls[cond4]
d=subset_d['total_actions']
temp=d.sum()
summation.append(temp)
temp2= subset_d['post_type'].unique()
x.append(temp2)

subset_e = df_mls[cond5]
e=subset_e['total_actions']
temp=e.sum()
summation.append(temp)
temp2= subset_e['post_type'].unique()
x.append('event')

for i in range(0,1):
    for j in range(0,len(summation)):
        plt.bar(x[j],summation[j], align='center', alpha=1)
        
    plt.xlabel('Post Type')
    plt.ylabel('Total actions')
    plt.show()
```

According to above bar graph we can very well conclude that maximum number of actions was received for the post type videp followed by links ,photo and status. Also there was no post for post type 'event' on the page Major league soccer(MLS)
